```python
response = requests.get(url)
```

```python
import json

response_data = json.loads(response.text)
data = response_data["collection"][0]["html"]
next_url = response_data["url"]
```

```python
next_url
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text)
print(soup)
```

```python
data = BeautifulSoup(data)
```

```python
data_list=[]

for d in data:
    title = d.find_all("span",{"class":"sds-comps-text sds-comps-text-ellipsis-1 sds-comps-text-type-headline1"})
    for t in title: #remove html text and only return the title
        data_list.append(t.get_text()) 
```

```python
data_list
```

```python
import json

clean_list = [title.strip() for title in data_list]

with open("title_chuncehon_news.json", 'w', encoding='utf-8') as f:
    json.dump(clean_list, f, ensure_ascii=False, indent=2)

```
